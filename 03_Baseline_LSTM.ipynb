# Baseline LSTM Model

In this notebook, we build our first Long Short-Term Memory (LSTM) model to predict daily household energy consumption.

This model will serve as a baseline for comparison with more advanced models later in the project.

# Section 2 - Import Libraries

```python
import numpy as np 
import  matplotlib.pyplot as plt

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM
from tensorflow.keras.layers import Dense
```

# Section 3 - Load Data

```python
X_train = np.load("X_train.npy")
X_test = np.load("X_test.npy")

y_train = np.load("y_train.npy")
y_test = np.load("y_test.npy")

print("Training:", X_train.shape)
print("Testing :", X_test.shape)
```

# Section 4 - model

```python
model = Sequential ([

    LSTM(

        units = 50 ,

        input_shape = (60,7)
    ) ,

    Dense(1)
])
```

```python
model.summary()
```

```python
model.compile (

    optimizer = 'adam' ,
    loss = "mse",
    metrics=["mae"]
)
```

# Section 5 — Train the Model

```python
history = model.fit(
    X_train,
    y_train,
    validation_data=(X_test, y_test),
    epochs=20,
    batch_size=32,
    verbose=1
)
```

```python
print(history.history.keys())
```

```python
print(history.history["loss"])
print(history.history["val_loss"])
```

## Section 6 — Plot Training History

```python
plt.figure(figsize=(10,5))

plt.plot(history.history["loss"], label="Training Loss")
plt.plot(history.history["val_loss"], label="Validation Loss")

plt.title("Training vs Validation Loss")
plt.xlabel("Epoch")
plt.ylabel("Loss")

plt.legend()

plt.grid(True)

plt.show()
```

```python
plt.figure(figsize=(10,5))

plt.plot(history.history["mae"], label="Training MAE")
plt.plot(history.history["val_mae"], label="Validation MAE")

plt.title("Training vs Validation MAE")
plt.xlabel("Epoch")
plt.ylabel("MAE")

plt.legend()

plt.grid(True)

plt.show()
```

# Section 7 — Make Predictions

```python
y_pred = model.predict(X_test)
```

# Section 8 — Model Evaluation

```python
from sklearn.metrics import (
    mean_absolute_error,
    mean_squared_error,
    r2_score
)

mae = mean_absolute_error(y_test, y_pred)
rmse = np.sqrt(mean_squared_error(y_test, y_pred))
r2 = r2_score(y_test, y_pred)

print(f"MAE  : {mae:.4f}")
print(f"RMSE : {rmse:.4f}")
print(f"R²   : {r2:.4f}")
```

# Section 9 — Actual vs Predicted

```python
plt.figure(figsize=(15,6))

plt.plot(y_test, label="Actual")
plt.plot(y_pred, label="Predicted")

plt.title("Actual vs Predicted Energy Consumption")

plt.xlabel("Days")
plt.ylabel("Power")

plt.legend()

plt.grid(True)

plt.show()
```
