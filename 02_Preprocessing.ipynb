# Data Preprocessing

In this notebook, we prepare the daily energy consumption data for training the LSTM model.

Steps:
- Load cleaned dataset
- Select features
- Normalize the data
- Create sliding windows
- Split into training and testing sets

## Section 2: Import Libraries

```python
import numpy as np
import pandas as pd

from sklearn.preprocessing import MinMaxScaler
from sklearn.model_selection import train_test_split
```

## Section 3: Load Clean Dataset

```python
daily_data = pd.read_csv(
    "daily_energy_data.csv",
    index_col=0,
    parse_dates=True
)

daily_data.head()
```

## Section 4: Check Dataset

```python
print(daily_data.shape)

daily_data.info()
```

## Section 5: Feature Selection

```python
features = [
    'Global_active_power',
    'Global_reactive_power',
    'Voltage',
    'Global_intensity',
    'Sub_metering_1',
    'Sub_metering_2',
    'Sub_metering_3'
]

data = daily_data[features]

data.head()
```

```python
daily_data = daily_data.dropna().reset_index(drop=True)
```

```python
print(len(daily_data))
```

```python
print(daily_data.isna().sum())
```

## Section 6: Normalization

```python
scaler = MinMaxScaler()
scaled_data = scaler.fit_transform(daily_data)
```

```python
print(np.isnan(scaled_data).sum())
```

## Section 7: Check Scaling

```python
print (scaled_data.shape)

scaled_data
```

## Section 8: Sliding Window

```python
window_size = 60

X = []
y = []

for i in range (window_size , len(scaled_data)) :

    X.append( scaled_data[i-window_size:i])
    
    y.append( scaled_data[i,0])
```

## Section 9: Convert to NumPy

```python
X = np.array(X)
y = np.array(y)
```

## Section 10: Check Shapes

```python
print(X.shape)

print (y.shape)
```

## Section 11: Train/Test Split

```python
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    test_size=0.2,
    shuffle=False
)
```

## Section 12: Check Shapes

```python
print(X_train.shape)
print(X_test.shape)

print(y_train.shape)
print(y_test.shape)
```

## Section 13: Save Prepared Data

```python
np.save("X_train.npy", X_train)
np.save("X_test.npy", X_test)

np.save("y_train.npy", y_train)
np.save("y_test.npy", y_test)
```
